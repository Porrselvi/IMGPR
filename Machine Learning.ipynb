```python
#Linearly Distributed Dataset Generation using Sklearn library
%matplotlib inline
from sklearn import datasets
from matplotlib import pyplot as plt
from sklearn.datasets.samples_generator import make_regression

X,y = make_regression(n_samples= 100, n_features = 1, noise = 5.4)
plt.scatter(X,y)

```

```python
#Make_blobs -> Cluster of random data. This is very useful for unsupervised learning.

import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import make_blobs


X,y = make_blobs(500, centers=3)

rgb = np.array(['r','g','b'])


plt.scatter(X[:,0],X[:,1],color = rgb[y])

#O/p comes randomly.
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import make_circles

X,y =  make_circles(n_samples=100, noise=0.09)
rgb = np.array(['r','g','b'])

plt.scatter(X[:,0],X[:,1],color =rgb[y])
```
