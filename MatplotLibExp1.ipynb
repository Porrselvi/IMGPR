```python
from matplotlib import pyplot as plt

rain = [4,5,6,3,8,9,4,6,2,7,5,6]
months = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']

plt.bar(months, rain, align='center', color='orange')
plt.xticks(months, rotation = 'vertical')
plt.show()
```

```python
import matplotlib.pyplot as plt
labels= ['LA','WA','CL','NY','FL']
corona = [24,56,45,67,77]
#To show separately, I am using explote attribute
explode = (0,0,0.2,0,0.3)
#Customizing colors
colors = ['green','red','blue','yellow']
plt.pie(corona, explode = explode, labels = labels, colors=colors, autopct = '%.1f%%', shadow = True,  startangle=0)
plt.axis('equal')
plt.title('Corona affected areas-2020')
plt.show()
```

```python
import matplotlib.pyplot as plt
labels= ['LA','WA','CL','NY','FL']
corona = [24,56,45,67,77]
#To show separately, I am using explote attribute
explode = (0,0,0,0,0)
#Customizing colors
colors = ['green','red','blue','yellow','lightblue']
plt.pie(corona, explode = explode, labels = labels, colors=colors, autopct = '%.1f%%', shadow = True,  startangle=0)
plt.axis('equal')
plt.title('Corona affected areas-2020')
plt.show()
```
